```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\Kriti Kapoor\OneDrive\Documents\GitHub\NLP_W2022\Datasets\SBIC.v2.agg.trn.csv') 

df = df[df.whoTarget.notna()]
#len(df.targetMinority.unique())
#df = df[df.targetMinority != '[]']
df.drop('dataSource', inplace=True, axis=1)
df.drop('offensiveYN', inplace=True, axis=1)
df.drop('sexYN', inplace=True, axis=1)
df.drop('targetStereotype', inplace=True, axis=1)
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
df['sentence'] = df['post']
df['sentence'] = df['sentence'].apply(lambda x: " ".join(x.lower() for x in x.split() if (x not in stop)))
df['sentence'] = df['sentence'].str.replace('[0-9]','')
df['sentence'] = df['sentence'].str.replace('@[\w]+','')
df['sentence'] = df['sentence'].str.replace('[^\w\s]','')
df['sentence'] = df['sentence'].str.replace(' rt ','')
df['sentence'] = df['sentence'].str.replace('[\n]','')
```

```python
df.drop('post', inplace=True, axis=1)
df.drop('whoTarget', inplace=True, axis=1)
df.drop('intentYN', inplace=True, axis=1)
df.drop('hasBiasedImplication', inplace=True, axis=1)
df.head()
```

```python


df.targetCategory = df.targetCategory.replace('"','', regex=True)
df.targetMinority = df.targetMinority.replace('"','', regex=True)
category = ['social','culture','race']
new_list=[]
newDf = pd.DataFrame(columns=df.columns)
df['targetCategory'] =  df['targetCategory'].apply(lambda x:x.replace('[','').replace(']',''))
df['targetMinority'] =  df['targetMinority'].apply(lambda x:x.replace('[','').replace(']','').lower())

for x in category:
    for item in df.targetCategory:
        if x in item:
            if item not in new_list:
              new_list.append(item)
newcol=[]
for item in df.targetCategory:
    if item in new_list:
     newcol.append(1)
    else:
        newcol.append(0)
df['newcol']=newcol
df = df[df.newcol == 1]
df.drop('newcol', inplace=True, axis=1)
df
```

```python
df.targetCategory = df.targetCategory.str.replace('race','racial')
```

```python


import itertools
liberal = ['antifa', 'liberal', 'anti-fa', 'antifascist', 'democrat', 'feminist', 'globalist', 'hilary clinton', 'left', 'socialist', 'social']
conservative = ['alt-right', 'conservative', 'libertarian', 'pro-life', 'republican', 'trump']

political = itertools.chain(liberal,conservative)
political = list(political)

political_list=[]
for x in political:
     for item in df.targetMinority:
        if x in item:
            if item not in political_list:
              political_list.append(item)
tempcol=[]
for item in df.targetMinority:
    if item in political_list:
     tempcol.append(1)
    else:
        tempcol.append(0)
df['tempcol']=tempcol
politicaldf = df[df.tempcol == 1]
politicaldf.drop('tempcol', inplace=True, axis=1)
df.drop('tempcol', inplace=True, axis=1)
politicaldf
```
