```python
import cPickle 

[lib, con, neutral] = cPickle.load(open('ibcData.pkl', 'rb'))
```

```python
liberal = []
for tree in lib:
    liberal.append(tree.get_words())
len(liberal)
```

```python
conservative = []
for tree in con:
    conservative.append(tree.get_words())
len(conservative)
```

```python
neut = []
for tree in neutral:
    neut.append(tree.get_words())
len(neut)
```

```python
liberalTag = [1] * len(liberal)
liberalTag
```

```python
conTag = [-1] * len(conservative)
conTag
```

```python
nTag = [0] * len(neut)
nTag
```

```python
sentences = liberal + conservative + neut
labels = liberalTag + conTag + nTag
```

```python
sentences[0]
```

```python
labels[0]
```

```python
import csv
from itertools import izip

with open('some.csv', 'wb') as f:
    writer = csv.writer(f)
    writer.writerows(izip(sentences, labels))
```
